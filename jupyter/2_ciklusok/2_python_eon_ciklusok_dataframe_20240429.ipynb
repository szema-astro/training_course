```python
import openpyxl
from openpyxl import load_workbook


# Ez a függvény egy Excel fájl teljes tartalmát beolvassa
# Minden fülön lévő adatot egy-egy külön DataFrame-be ment el
# A DataFrame-eket egy listában tárolja. Ha volt az Excel-ben 5 fül, akkor 5 DataFrame lesz, és ezeket egy listába teszi be
# Azaz a függvény visszaad egy olyan listát, ami DataFrame-eket tartalmaz!

def read_excel_file(f_name):
    wb = load_workbook(f_name)
    sheets = wb.sheetnames
    df_list = []
    for sheet in sheets:
        ws = wb[sheet]
        data = ws.values
        # Get the first line in file as a header line
        columns = next(data)[0:]
        # Create a DataFrame based on the second and subsequent lines of data
        df = pd.DataFrame(data, columns=columns)
        df_list.append(df)
    return df_list
```

# I.2.3. enumerate() függvény

Az enumerate() függvény akkor használjuk, amikor a for ciklus egy lista elemein megyünk végig. Ekkor két mozgó változónk van:
- első mozgó változó (i): egy numerikus változó, ami azt mutatja hogy hanyadik ciklusban vagyunk, pontosabban a lista hanyadik eleme került a másik mozgó változóba (Python-ban megszokott módon 0-val indul, és egyesével növekszik az értéke)
- második mozgó változó (col): ebbe kerülnek a ciklus futása során a lista elemei

```python
gyumolcsok = ["alma", "körte", "banán", "eper", "dió", "narancs"]

for i, col in enumerate(gyumolcsok):
    print("i értéke:", i, "             col értéke:", col)
```

# I.2.4. iterrows() függvény

- az első mozgó változó itt is számolja hányadik ciklusban vagyunk
- a második mozgó változóba kerül a DataFrame teljes sora (ha 25 soros a DatFrame, akkor a for ciklus 25-ször megy végig, és minden ciklusban a DataFrame egy-egy sora kerül ebbe a mozgó változóba)

```python
# try utáni utasításokat megpróbálja végrehajtani a fordító
# ha hibára fut, akkor nem áll le a furás, hanem az except ágon folytatja

try:
    print(x)  # nem definiáltuk az x változót!
except:
    print("Valami probléma volt")
```

```python
# Vannak előredefiniált hibák
# ZeroDivisionError, NameError, TypeError, ValueError, RuntimeError, ...

try:
    print(x)
except NameError:
    print("x változó nincs definiálva")
except:
    print("Valami más a probléma")
```

```python
# try utasításon belüli for cíiklus nem megy végig

x = ["a","b"] 
try:
    for i in range(4):
        print(i, x[i])
except:
    print(i, "Ilyen indexű lista érték nincs")
```

# IV.1. Aggregálás

Az aggregálás során bizonyos változókra kiszámoljuk más változók aggregált értékeit (pl. átlagértéket vagy minimum értéket). Az aggregálást a groupby() és az agg() függvények kombinációjával tudjuk előállítani.  

```python
df1_filter = df1[list(df1.columns)[:7]].copy()
df1_filter
```

```python
df1_filter.columns = ["StartDate", "EndDate", "Duration", "FEEDBACK", "SCORE", "ID", "UP"]
df1_filter = df1_filter[["ID", "StartDate", "EndDate", "Duration", "FEEDBACK", "SCORE",  "UP"]]
```

```python
df1_filter
```

```python
# Az alábbi sor egy aggregálást hajt végre:
# df1.groupby() függvényben a df1 egy DataFrame neve
# a groupby(["field1", "field2", ...], as_index=False) esetében a ["field1", "field2", ...] rész mondja meg, 
# hogy mely változókra aggregáljunk, az  as_index=False azt érjük el, hogy az aggregált ne kerüljenek az aggregálás után
# az index mezőbe. 
# az groupby() függvényt az agg(dict) függvény követi, ahol a dict írja le mely mezőkkel milyen aggregálást hajtsunk végre
# {field: ["min", "max", "mean", "median", "sum", "sdev"]} az adott mező min, max .. értékeit számolja ki! 
 
df1_agg = df1_filter.groupby(["FEEDBACK"], as_index=False).agg({"Duration": ["min", "max", "mean", "count"]})
df1_agg
```

```python
# Aggregálás után a mezőneveket érdemes átírni
# Legegyszerűbb módja, hogy a column listát felülírjuk egy új listával

df1_agg.columns = ["FEEDBACK","duration_min", "duration_max", "duration_atlag", "count"]
df1_agg
```

```python
df1_agg["duration_atlag"] = df1_agg["duration_atlag"].astype("int64")
df1_agg
```

```python
# Már tanultuk, hogy a value_counts() függvénnyel visszakpjuk egy változóra értékeinek eloszlását. 

df1_filter["FEEDBACK"].value_counts()
```

```python
# Alapértelmezés szerint az üres értékek darabszámát nem kapjuk meg, de a (dropna=False) paraméterrel ez is megjeleníthető 

df1_filter["FEEDBACK"].value_counts(dropna=False)
```

```python
# Az eloszlást aggregálással is elő lehet állítanui - DE
# "Duration" helyett itt bármilyen más mező nevet lehet írni! 

df1_1_agg = df1_filter.groupby(["FEEDBACK"], as_index=False).agg({"Duration":  "count"})
df1_1_agg
```

# VI. Táblák összekapcsolás és összefűzése

Nagyon sokszor egyszerre több táblával kell dolgozni. Ilyenkor többféle módon lehet ezeket összekapcsolni: kulcs vagy kulcsok alapján vagy csak egyszerűen összefűzni. Első esetben merge utasítást használjuk, a második esetben a concat utasítást. 

## VI.1. Táblák összekapcsolása

```python
df1 = pd.DataFrame({"fruit": ["alma", "körte", "banán"], "price": [200, 330, 500]})
```

```python
df1
```

```python
df2 = pd.DataFrame({"fruit": ["alma", "eper", "körte"], "discount": [2, 10, 3]})
```

```python
df2
```

```python
df3 = df2.copy()
df3.columns = ["gyumi", "kedvezmeny"]
df3
```

```python
# Két DataFrame-et a pd.merge() függvénnyel tudunk összekapcsolni. 
# Az első két paraméter mindig a két DataFrame neve
# on= paraméterben adjuk meg a kulcsmező nevét, ami alapján összekapcsoljuk a két táblát
# how paraméter értéke lehet: "inner", "outer", "right", "left" illetve "cross" - ami a Descartes szorzatot adja a két táblának

df_inner = pd.merge(df1,df2, on="fruit", how="inner")
df_inner
```

```python
# Akkor is összetudunk kapcsolni két táblát, ha a kulcsmezők nevei nem egyeznek
# Ilyenkor a left_on és right_on paraméterben kell megadni a kulcsmezők neveit!
# Ilyenkor viszont mindkét kulcsmező bekerül az eredmény táblába

df_inner2 = pd.merge(df1,df3, left_on="fruit", right_on="gyumi", how="inner")
df_inner2
```

```python
# Ha a két táblában a kulcsmezők nevei megegyeznek, akkor az on paraméterben kell megadni a kulcsmező nevét

df_inner = pd.merge(df1,df2, on="fruit", how="inner")
df_inner
```

```python
df_outer = pd.merge(df1,df2, on="fruit", how="outer")
df_outer
```

```python
df_left = pd.merge(df1,df2, on="fruit", how="left")
df_left
```

# VI.2. Táblák összefűzése

```python
df1 = pd.DataFrame({"fruit": ["alma", "körte", "banán"], "price": [200, 330, 500], "discount":[2,2,12]})
```

```python
df1
```

```python
df2 = pd.DataFrame({"fruit": ["dinnye", "eper", "dió"], "price": [760, 450, 1100], "discount":[1,20,12]})
```

```python
df2
```

```python
# az axis = 0 paraméter miat a concat a két táblát "egymás alá" helyezi

df_concat_függőleges = pd.concat([df1,df2], axis = 0)
df_concat_függőleges
```

```python
df3 = df2.copy()
df3["uj"] = 0
df3
```

```python
# A concat() nem "érzékeny" arra, hogy a két tábla struktúrája mennyire egyezik meg. 
# Ha vannk eltérő mezők, akkor a mezőlista uniója kerül be az eredmény táblába
# Amely mező nem volt egyik táblában, annak a tábla sorainál üres értékek lesznek

d_teszt_concat_df = pd.concat([df1, df3], axis = 0)
d_teszt_concat_df
```

```python
# Ha az indexeket újra szeretnénk kiosztani, akkor az "ignore_index"  paramétert True értékre kell állítani

df_concat_függőleges = pd.concat([df1,df2], axis = 0, ignore_index=True)
df_concat_függőleges
```
